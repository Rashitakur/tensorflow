```python
# Clean up column names by stripping any unwanted spaces or tab characters
df.columns = df.columns.str.strip()

# Define the list of columns
yes_no_columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
                  'PhoneService', 'MultipleLines', 'InternetService',
                  'OnlineSecurity', 'DeviceProtection', 'TechSupport', 'StreamingTV',
                  'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 
                  'MonthlyCharges', 'TotalCharges', 'Churn']

# Replace 'Yes' with 1 and 'No' with 0 for each column in the list
for col in yes_no_columns:
    if col in df.columns:
        df[col] = df[col].replace({'Yes': 1, 'No': 0})
    else:
        print(f"Column {col} not found in DataFrame")
```

```python
for col in df:
    print(f'{col}:{df[col].unique()}')
```

dummies variable for DSL and Fibre optic (cleaning)

```python
df=pd.get_dummies(data=df,columns=['InternetService','PaymentMethod','Contract'])
```

```python
df
```

```python
# Create a mapping for the necessary conversions
conversion_dict = {
    'MultipleLines': {'NO': 0, 0: 0, 1: 1},
    #'InternetService': {'DSL': 1, 'Fiber optic': 2, 0: 0},
    'OnlineSecurity': {'NO': 0, 0: 0, 1: 1},
    'OnlineBackup': {'Yes': 1, 'No': 0, 'NO': 0},
    'DeviceProtection': {'NO': 0, 0: 0, 1: 1},
    'TechSupport': {'NO': 0, 0: 0, 1: 1},
    'StreamingTV': {'NO': 0, 0: 0, 1: 1},
    'StreamingMovies': {'NO': 0, 0: 0, 1: 1}
}

# Loop through the columns and apply the replacement using the conversion_dict
for col, mapping in conversion_dict.items():
    if col in df.columns:
        df[col] = df[col].replace(mapping)
    else:
        print(f"Column {col} not found in DataFrame")
```

```python
# Round the values to the nearest integer before converting
df['MonthlyCharges'] = df['MonthlyCharges'].round().astype(int)
df['TotalCharges'] = df['TotalCharges'].round().astype(int)
# Convert 'MonthlyCharges' and 'TotalCharges' to integer type
df['MonthlyCharges'] = df['MonthlyCharges'].astype(int)
df['TotalCharges'] = df['TotalCharges'].astype(int)
```

```python
df.dtypes
```

```python
cols_to_scale=['tenure','MonthlyCharges','TotalCharges']
from sklearn. preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[cols_to_scale]=scaler.fit_transform(df[cols_to_scale])
```

```python
df.sample(4)
```

```python
X=df.drop('Churn',axis='columns')
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(5, input_shape=(26,), activation='relu'),
    keras.layers.Dense(100,  activation='sigmoid')
])

model.compile(
    optimizer="adam",
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_test[:5]
```

```python
yp=model.predict(X_test)
yp[:5]
```

```python
# Assuming yp is a numpy array, use vectorized operations
import numpy as np

y_predict = (np.array(yp) > 0.5).astype(int).tolist()
y_predict
```

```python
y_predict[:5]
```

```python
import tensorflow as tf
from sklearn.metrics import classification_report

# Assuming y_test and y_predict are your true and predicted labels
# Check the shape of y_predict
print(y_predict.shape)

# If y_predict is a 2D tensor, apply tf.argmax to get the predicted class
if len(y_predict.shape) > 1:
    y_predict = tf.argmax(y_predict, axis=1)

# Now you can generate the classification report
report = classification_report(y_test, y_predict)
print(report)
```

```python

import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
if len(y_predict.shape) > 1:
    # Convert logits/probabilities to predicted class labels
    y_predict = tf.argmax(y_predict, axis=1)

# Now compute the confusion matrix
cm = confusion_matrix(y_test, y_predict)

# Convert to integers for display (in case values are in scientific notation)
cm = cm.astype(int)

# Plot the confusion matrix
plt.figure(figsize=(8, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')  # 'fmt="d"' ensures integers
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
